```python
#Tính giai thừa của số nguyên dương n
#Dùng vòng lặp
n = int(input("Nhập n: "))
result = 1
for i in range(1, n+1):
    result = result * i
print(f"5! = {result}")
```

```python
#Tính giai thừa của số nguyên dương n
#Dùng đệ quy
n = int(input("Nhập n: "))
def facto(n):
    if n == 0:
        return 1
    else:
        return n * facto(n-1)
print(f"5! = {facto(n)}")
```

```python
#Tính tổ hợp chập k của n phần tử
#Dùng hàm facto đã tạo ở trên
n,k = map(int, input("Nhập n,k: ").split())
Cnk = facto(n) // (facto(k) * facto(n-k))
print(f"C({n},{k}) = {Cnk}")
```

```python
#Tính số thứ n trong dãy Fibonacci
#Dùng đệ quy
n = int(input("Nhập n: "))
def fibo(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibo(n-1) + fibo(n-2)
print("Kết quả: ", fibo(n))
```

```python
#Tính số thứ n trong dãy Fibonacci
#Dùng list và vòng lặp
def fibo_2(n):
    F = [0, 1]
    for i in range(2, n + 1):
        x = F[i - 2] + F[i - 1]
        F.append(x)
    return F[n]
print(fibo_2(4))
```

```python
#Tính số thứ n trong dãy Fibonacci
#Dùng list và vòng lặp (tối ưu bộ nhớ)
def fibo_3(n):
    F = [0,1,1]
    for i in range(2, n+1):
        F[2] = F[0] + F[1]
        F[0] = F[1]
        F[1] = F[2]
    return F[1]
print(fibo_3(10**4))
```
