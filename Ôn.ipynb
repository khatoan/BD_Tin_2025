```python
K = 10
T = 0
x = 0
while T < K:
    x += 1
    T += 2**x
T -= 2**x
print("Giá trị của x:", x)
print("Tổng T:", T)
stt = K - T - 1
print(stt)
stt = bin(stt)[2:]
print("Số nhị phân:", stt)
for i in range(x - len(stt)):
    stt = '0' + stt
print("Số nhị phân sau khi thêm:", stt)
s = ''
for i in stt:
    if i == "0":
        s += "5"
    if i == "1":
        s += "8"
print(s)
```
