```python
#Tìm kiếm nhị phân
#Dùng vòng lặp
A = [2, 4, 5, 7, 9, 11, 15, 17, 19]
x = int(input("Nhập phần tử cần tìm:"))
t = 0
p = len(A)-1
kq = -1
while t< p and kq == -1:
    mid = (t+p)//2
    if A[mid]==x:
        kq = mid
    elif A[mid] <x:
        t = mid + 1
    else:
        p = mid - 1
print("Vị trí của phần tử cần tìm là: ", kq)
```

```python
#Tìm kiếm nhị phân
#Dùng đệ quy
def binary_search(A, t, p, x):
    mid = t + (p - t)//2
    if A[mid] == x:
        return mid
    elif A[mid] < x:
        return binary_search(A, mid + 1, p, x)
    else:
        return binary_search(A, t, mid - 1, x)
A = [2, 4, 5, 7, 9, 11, 15, 17, 19]
x = 9 
kq = binary_search(A, 0, len(A)-1, x)
print("Vị trí của phần tử cần tìm là: ", kq)
```

```python
#Hàm tính lũy thừa
#Dùng phương pháp chia để trị
def lthua(a, n):
    if n == 0:
        return 1
    elif n % 2 == 0:
        return lthua(a, n//2) *lthua(a, n//2)
    else:
        return a * lthua(a, n-1)
print(lthua(2, 5))
```
