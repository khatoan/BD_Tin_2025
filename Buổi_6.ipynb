```python
#Ghép hai mảng đã sắp xếp và trả về mảng đã sắp xếp
def merge(A, T, P):
    i = j = k = 0
    while i < len(T) and j < len(P):
        if T[i] < P[j]:
            A[k] = T[i]
            i += 1
        else:
            A[k] = P[j]
            j += 1
        k += 1
    while i < len(T):
        A[k] = T[i]
        i += 1
        k += 1
    while j < len(P):
        A[k] = P[j]
        j += 1
        k += 1
    return A
```

```python
# Thuật toán sắp xếp trộn (Merge Sort)
def Merge_sort(arr):
    if len(arr) <= 1:
        return arr
    mid = len(arr)//2
    T = arr[:mid]
    P = arr[mid:]
    Merge_sort(T)
    Merge_sort(P)
    # Gọi hàm merge đã viết ở trên
    merge(arr, T, P)
    return arr
```

```python
# Vận dụng/45
A = [3, 7, 2, 6]
B = [4, 2, 8, 5]
A = Merge_sort(A)
B = Merge_sort(B)
print("Các cặp xếp theo yêu cầu: ")
for i in range(len(A)):
    print(f"({A[i]}, {B[i]})")
```
