```python
# Tìm phần tử lớn nhất trong một mảng có phần đầu được sắp xếp tăng dần và phần sau được sắp xếp giảm dần.
# Dùng tìm kiếm nhị phân bằng đệ quy
def find_max(arr, left, right):
    k = (left + right) // 2
    if A[k] > A[k - 1] and A[k] > A[k + 1]:
        return A[k]
    elif A[k] > A[k - 1] and A[k] < A[k + 1]:
        return find_max(arr, k+1, right)
    else:
        return find_max(arr, left, k-1)
A = [1, 3, 8, 12, 4, 2]
right = len(A)-1
left = 0
print(find_max(A, left, right))
```

```python
#Tìm phần tử lớn nhất trong một mảng có phần đầu được sắp xếp tăng dần và phần sau được sắp xếp giảm dần.
# Dùng brute_force
def find_max_2(arr):
    for i in range(len(arr)-1):
        if arr[i] > arr[i+1] and arr[i] > arr[i-1]:
            return arr[i]
A = [1, 3, 8, 12, 4, 2]
print(find_max_2(A))
```

```python
#Ghép hai mảng đã sắp xếp và trả về mảng đã sắp xếp
def merge(A, T, P):
    i = j = k = 0
    while i < len(T) and j < len(P):
        if T[i] < P[j]:
            A[k] = T[i]
            i += 1
        else:
            A[k] = P[j]
            j += 1
        k += 1
    while i < len(T):
        A[k] = T[i]
        i += 1
        k += 1
    while j < len(P):
        A[k] = P[j]
        j += 1
        k += 1
    return A
T = [2, 5, 8]
P = [3, 6, 7, 9]
A = T + P
print(merge(A, T, P))
```

```python
# Thuật toán sắp xếp trộn (Merge Sort)
def Merge_sort(arr):
    if len(arr) <= 1:
        return arr
    mid = len(arr)//2
    T = arr[:mid]
    P = arr[mid:]
    Merge_sort(T)
    Merge_sort(P)
    # Gọi hàm merge đã viết ở trên
    merge(arr, T, P)
    return arr
arr = [38, 27, 43, 3, 9, 82, 10]
print(Merge_sort(arr))
```
