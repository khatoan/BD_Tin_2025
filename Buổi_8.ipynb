```python
#Liệt kê dãy bit độ dài n
#Dùng vòng lặp
n = 3 #độ dài dãy bit
s1 = [0 for i in range(n)]
s2 = [1 for i in range(n)]
print(s1)
while s1 != s2:
    i = n - 1
    while i > 0 and s1[i] == 1:
        i -= 1
    if s1[i] == 0:
        s1 = s1[:i] + [1] + [0 for i in range(i, n-1)] #Lặp từ i đến n-1 vì trừ đi vị trí [i] đã đổi thành 1
    print(s1)
```

```python
# Bài toán mua số đồ dùng trong n đồ dùng sao cho độ thích cao nhất và giá ko vượt quá T
# Dùng vòng lặp
T = 20 # Số tiền đang có
P = [10 , 5, 18, 9, 5] # Dãy các giá tiền
n = len(P) # Số món hàng
S = [7, 2, 12, 6, 3] # Dãy độ thích
best_solution = [] # Kết quả chọn cuối cùng ( biểu diễn dưới dạng dãy bit)
best_S = 0 #Độ thích lớn nhất
s_1 = [0 for i in range(n)]
s_2 = [1 for _ in range(n)]
while s_1 != s_2:
    i = len(s_1) - 1
    while i > 0 and s_1[i] == 1:
        i = i - 1
    if s_1[i] == 0:
        s_1 = s_1[:i] + [1] +  [0 for _ in range(i, n-1)]
    Gia = Thich = 0
    for k in range(n):
        if s_1 [k] == 1:
            Gia  = Gia + P[k]
            Thich = Thich + S[k]
            if Gia <= T and Thich > best_S:
                best_solution = s_1
                best_S = Thich
print("KQ chọn:", best_S, best_solution)
        
```
