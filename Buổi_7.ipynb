```python
# Chọn mua 1 sản phẩm
import random
T = int(input("Nhập số tiền bạn có:"))
P = list(map(int, input("Nhập giá các sản phẩm:").split())) # Cách đều bởi dấu cách
n = len(P)
S = list(map(int, input("Nhập mức độ yêu thích của các sản phẩm:").split())) # Cách đều bởi dấu cách
"""
P = [random.randint(1, 100) for _ in range(1000)]
print("Giá các sản phẩm:", P)
S = [random.randint(1, 10) for _ in range(1000)]
print("Mức độ yêu thích của các sản phẩm:", S)
"""
k = -1 
for i in range(n):
    if P[i] <= T:
        if k == -1 or S[i] > S[k]:
            k = i
if k == -1:
    print("Không mua được sản phẩm nào")
else:
    print("Mua 1 sản phẩm tại vị trí", k, "với giá", P[k])
```

```python
# Chọn mua 2 sản phẩm
T = 1000
P = [random.randint(1, 100) for i in range(1000)] # Giá các sản phẩm
n = len(P)
S = [random.randint(1, 100) for i in range(1000)] # Mức độ yêu thích của các sản phẩm
k_1 = k_2 = -1
for i in range(n+1):
    for j in range(i+1, n):
        if P[i] + P[j] <= T:
            if k_1 == -1 or k_2 == -1 or S[i] + S[j] > S[k_1] + S[k_2]:
                k_1, k_2 = i, j
if k_1 == -1 or k_2 == -1:
    print("Không mua được sản phẩm nào")
else:
    print("Mua 2 sản phẩm tại vị trí", k_1, "và", k_2, "với giá", P[k_1], "và", P[k_2])
```

```python
# Tìm nghiệm nguyên phương trình ax^5 + bx + c = 0
# Biết nghiệm là ước của c
a, b, c = 1, -15, -2
for i in range(1, abs(c)+1):
    if (a*i**5 + b*i + c == 0):
        print("Nghiệm là", i)
        break
    elif (a*(-i)**5 + b*(-i) + c == 0):
        print("Nghiệm là", -i)
        break 
```

```python
# Vận dụng trang 54
B = [170, 175, 180, 185]
G = [165, 168, 171, 173]
n = len(B)
x = 1000
y = 0
for i in range(n):
    for j in range(n):
        if (B[i] > G[j]) and (B[i] - G[j] < x - y):
            x = B[i]
            y = G[j]

if x == 1000 and y == 0:
    print("Không có cặp nam nữ thỏa mãn")
else:
    print("Cặp nam nữ có chiều cao gần nhau nhất là:")
    for i in range(n):
        for j in range(n):
            if (B[i] > G[j]) and (B[i] - G[j] == x - y):
                print(B[i], G[j])
```
