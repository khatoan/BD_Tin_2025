```python
# Bài toán 3: Số siêu nguyên tố
def nguyen_to(t):
    if t < 2:
        return False
    elif t == 2:
        return True
    else:
        for k in range(3, int(t**(1/2)+1)):
            if t % k == 0:
                return False
    return True
def backtrack(num, n):
    if len (str(num)) == n:
        print(num, end = " ")
    for v in [1,3,7, 9]:
        new_num = num * 10 + v
        if nguyen_to(new_num):
            backtrack(new_num, n)
def sieu_nguyen_to(n):
    for start in [2, 3, 5, 7]:
        backtrack(start, n)
n = 7 # Nhập độ dài n của số siêu nguyên tố
print("Các số siêu nguyên tố có", n, "chứ số")
sieu_nguyen_to(n)
```
