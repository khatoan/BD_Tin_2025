```python
#Hàm tính số cái bắt tay trong trong một nhóm n người
n = int(input("Nhập số người trong phòng: "))
def h(n):
    if n == 0 or n == 1:
        return 0
    else:
        return h(n-1) + (n-1)
print("Số cái bắt tay:", h(n))
```

```python
# Bản cải tiến 
n = int(input("Nhập số người trong phòng: "))
KQ = 0
if n >1:
    k = 2
    while k < n +1:
        KQ = KQ + k - 1
        k += 1
print("Số cái bắt tay:", KQ)
```

```python
# Hàm tính ước chung lớn nhất của 2 số nguyên dương
x, y  = map(int, input("Nhập 2 số nguyên dương trên cùng 1 hàng: ").split())
def UCLN(x, y):
    r = x % y
    if r == 0: 
        return y
    else: 
        return UCLN(y, r)
print("UCLN:", UCLN(x, y))
```

```python
# Trang 18 chuyên đề 11
def A(i , j, D, C):
    if(i == D or j == C):
        return 1
    else:
        return A(i + 1, j, D, C) + A(i, j + 1, D, C)
i , j, D, C = map(int, input("Nhập 4 số nguyên dương trên cùng 1 hàng: ").split())
print("Số cách đi từ (i,j) đến (D,C):", A(i, j, D, C))
```
