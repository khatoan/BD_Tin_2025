```python
#Tìm số gạch tối thiểu để lát sàn nhà có kích thước D x R
def F(D, R):
    if D <= 0 or R <= 0:
        return 0
    else: 
        i = 8
        SizeMin = min(D, R)
        while 2**i > SizeMin:
            i -=  1
        # u là số lượng viên gach cỡ 2**i có thể chứa trên cạnh D
        # v là số lượng viên gạch cỡ 2**i có thể chứa trên cạnh R
        u = D//2**i
        v = R//2**i
        return u*v + F(D - u*2**i, R) + F(u*2**i, R - v*2**i)
D, R = map(int, input("Nhập chiều dài, rộng cần lát: ").split()) 
print("Số lượng viên gạch tối thiểu cần lát:",F(D, R))   
```

```python
#Bài toán tháp Hà Nội 
def HN_tower(n, C1, C2, C3 ):
    if n == 1:
        print(C1, "-->", C2)
    else:
        HN_tower(n-1, C1, C3, C2)
        HN_tower(1, C1, C2, C3)
        HN_tower(n-1, C3, C2, C1)
n = int(input("Nhập số đĩa: "))
print("Số thao tác:", 2**n-1)
print("Các bước cụ thể:")
HN_tower(n, 'A', 'B', 'C')
```

```python
# Tính tổng chữ số của một số nguyên dương
#Dùng đệ quy
def sumDigits(n):
    if n < 10:
        return n
    else:
        return n%10 + sumDigits(n//10)
print("Tổng chữ số của n là:", sumDigits(12345))
```

```python
# Hàm đảo ngược một chuỗi bằng đệ quy
def reverseString(s):
    if len(s) == 1:
        return s
    else:
        return reverseString(s[1:]) + s[0] 
string = input("Nhập chuỗi cần đảo ngược: ")
print("Chuỗi đảo ngược là:", reverseString(string))
```
